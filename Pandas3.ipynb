# 10. fillna()

By “isnull()” and “sum()” checks the missing values in dataset.
we can see that this “driver_gender” feature has 5335 nan or missing values. 

So, we will fill up the missing values using the:
##### mode(a value that appears most frequently in a data set) of this particular feature using the “fillna()” function.

df["driver_gender"].isnull().sum()

```python
df["driver_gender"].value_counts()
```

```python
df["driver_gender"].fillna(df["driver_gender"].mode,inplace=True)
```

```python
df["driver_gender"].isnull().sum()
```

###### Note – This is NOT THE most effective way of filling missing/NAN values. 
###### There are other much more efficient ways for imputation. There is a lot of thought which goes behind when we are working with missing values. But, this is just for explaining the functionality of the fillna() function.

# get_dummies()
- get_dummies() method converts the categorical features of the data into dummy variables or indicator variables
- We usually do this conversion because some machine learning models don’t work well with categorical values such as Random Forrest, but we shouldn’t be using this if we have too many categories in our feature as it will create that many new features in our data frame, which will have an effect on the performance of our model.

#####   get_dummies()  method in Pandas provides a convenient way to perform One-Hot Encoding on data. 

```python
df[:3]
```

```python
df1 = pd.get_dummies(df.driver_race, prefix = 'driver_race_').iloc[:,0:]
```

# select_dtypes()
- Separate NUMARICAL and CATEGORICAL features from our data frame
- We can Create new ones by using the “select_dtypes()” function and include  [np.number]  to select numerical columns
- And include “objects” for categorical columns.

```python
df2= df.select_dtypes(include=[np.number])

df2[4:8]
```

```python
df3 = df.select_dtypes(include='object')

df3[:5]
```

# concat()
- We can perform concatenation of pandas object into a DataFrame output along a particular axis with optional set logic such as union and intersection using concat() method.

- By default, axis=0, i.e. row-wise concatenation, so if we set axis=1, column-wise concatenation will be performed.

```python
df_combo = pd.concat([df2,df3],axis=1)
```

```python
df_combo[:3]
```

# apply()
- while applying self defined function to your dataframe
- apply() :allows us to apply a custom function to every element of a particular Series.

```python
def CurrentAgeCalculator(dob):
    return (2023-dob)


df['driver_current_age'] = df['driver_age_raw'].apply(CurrentAgeCalculator)
```
